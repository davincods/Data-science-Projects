```python
# importing the libraries again
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("austin_weather_final.csv")
```

```python
data
```

```python
X= data.drop(['PrecipitationSumInches'], axis=1)
```

```python
Y= data["PrecipitationSumInches"]
```

```python
# reshaping into the 2d vector
Y=Y.values.reshape(-1,1)
```

```python
days_index=798
days= [i for i in range(Y.size)]
```

```python
# Initialize the linear regression classifier
clf= LinearRegression()

#Train the classifier
clf.fit(X,Y)
```

```python
# Plot the graph now
print("The precipitation graph looks like :")
plt.scatter(days,Y, color= "g")
plt.scatter(days[days_index],Y[days_index], color= "r")
plt.title("Precipitation level")
plt.xlabel("Days")
plt.ylabel("Precipition in inches")
plt.show()
x_vis= X.filter(["TempAvgF","DewPointAvgF","HumidityAvgPercent", "SeaLevelPressureAvgInches","VisibilityAvgMiles","WindAvgMPH"])
```

```python
# To plot the graph b/w the precipitation and the attributes
print (" The graph b/w the precipitation and the attributes")

for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days, x_vis[x_vis.columns.values[i][:100]], color="g")
    plt.scatter(days[days_index], x_vis.iloc[days_index, i], color="r", label='Precipitation > 0')
    plt.title(x_vis.columns.values[i])
    plt.xlabel("Days")
    plt.ylabel(x_vis.columns[i])
    
plt.show()
```
